```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = 'Денситометр'
dataframe = pd.read_csv(filename)

x = [1, 109]
y = [822.8584, 1835.5243]

fig, ax = plt.subplots()
ax.plot(dataframe["Итерация"], dataframe["Денситометр"], label ='Данные')
ax.plot(x,y,label = 'Вектор предсказания')
ax.set_title('Денситометр')
ax.legend()
fig.show()
```
